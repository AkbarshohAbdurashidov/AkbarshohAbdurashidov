### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

```python
path='https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
df=pd.read_csv(path,sep='\t')
df.head()
```

### Step 3. Assign it to a variable called chipo.

```python
df.info()
df['item_price'] = df['item_price'].apply(lambda x: float(x[1:]))
df.head()
```

### Step 4. How many products cost more than $10.00?

```python
mask=df['item_price'] > 10
df[mask]
print(len(df[mask]), 'ta buyurtma narxi 10$ dan qimmatroq.')
```

### Step 5. What is the price of each item?
###### print a data frame with only two columns item_name and item_price

```python
df[['item_name','item_price']]

```

### Step 6. Sort by the name of the item

```python
df.sort_values(by='item_name')
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
#method 1
df.sort_values(by=['item_price'], ascending=False).head(1)
# method 2
df.sort_values(by=['item_price']).tail(1)
# idxmax = index of max
max_id = df['item_price'].idxmax()
max_id
df.loc[[max_id]]
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
mask = df['item_name'] == 'Veggie Salad Bowl'
df[mask]
print(len(df[mask]), 'ta Veggie Salad Bowl buyurtma berildi.')
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
mask  = (df['item_name'] == 'Canned Soda') & (df['quantity'] > 1)
df[mask]
print(len(df[mask]), 'ta Canned Soda buyurtma berildi.')
```
